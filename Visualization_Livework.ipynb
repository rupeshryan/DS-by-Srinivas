# Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

### UniVariate - "one variable" data visualization

```python
#Strip plot
sns.stripplot(y= 'tip', data= tips, jitter= True)  #jitter will prevent the datat points to be in single line so that we can see better
plt.show()
```

```python
#grouping with strip plot
sns.stripplot(x='day', y='tip', data=tips)
plt.show()
```

```python
sns.stripplot(x= "day", y="tip",data=tips, size=4, jitter=False)
plt.show()
```

```python
#Swarm Plot
sns.swarmplot(x='day',y='tip', data= tips)
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')    #here we are adding sex as another variable to differentiate tips based on gender
plt.show()
```

```python
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h')    #changing orientation
plt.show()
```

```python
 #Box Plot
 sns.boxplot(x='day', y='tip', data= tips)
 plt.show()
```

```python
#Violin plot
sns.violinplot(x='day', y='tip', data= tips)
plt.show()
```

```python
#putting the plots side by side
plt.subplot(1,2,1)    #1,2 is partition and 1 is position
sns.boxplot(x='day', y='tip', data= tips)
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data= tips)
plt.show()
```

```python
#cominbing the plot
sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.show()
```

## Bi-variate - two variable visualition

```python
#Joint plot
sns.jointplot(x='total_bill', y= 'tip', data=tips)
plt.show()
```

```python
#density plot
sns.jointplot(x= 'total_bill', y='tip', data= tips, kind = 'kde')    #kde  is kernal density estimation
plt.show()
```

```python
#pair plot
sns.pairplot(tips)
```

```python
sns.pairplot(tips , hue='sex')
```
